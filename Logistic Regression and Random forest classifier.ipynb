```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv(" ai_assistant_usage_student_life.csv")
```

```python
df.head(1)
```

```python
df.isnull().sum()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
lb=LabelEncoder()
```

```python
df["SessionID"]=lb.fit_transform(df["SessionID"])
df["StudentLevel"]=lb.fit_transform(df["StudentLevel"])
df["Discipline"]=lb.fit_transform(df["Discipline"])
df["TaskType"]=lb.fit_transform(df["TaskType"])
df["FinalOutcome"]=lb.fit_transform(df["FinalOutcome"])
df["UsedAgain"]=lb.fit_transform(df["UsedAgain"])
df["SessionDate"]=lb.fit_transform(df["SessionDate"])
```

```python
df.head(1)
```

```python
x=df.drop(columns=["UsedAgain"])
y=df["UsedAgain"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

 # LogisticRegression

```python
from sklearn.linear_model import  LogisticRegression
```

```python
lr=LogisticRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
y_pred=lr.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_pred,y_test)
```

# RandomForestClassifier

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf=RandomForestClassifier()
```

```python
rf.fit(x_train,y_train)
```

```python
y_pred=rf.predict(x_test)
```

```python
accuracy_score(y_pred,y_test)
```
