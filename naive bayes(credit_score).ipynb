```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("credit_scoring.csv")
```

```python
df.head(1)
```

```python
df.isnull().sum()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
lb=LabelEncoder()
```

```python
df.head(1)
```

```python
df["Gender"]=lb.fit_transform(df["Gender"])
df["Marital Status"]=lb.fit_transform(df["Marital Status"])
df["Employment Status"]=lb.fit_transform(df["Employment Status"])
df["Education Level"]=lb.fit_transform(df["Education Level"])
df["Type of Loan"]=lb.fit_transform(df["Type of Loan"])

```

```python
df.head(1)
```

```python
x=df.drop(columns=["Type of Loan"])
y=df["Type of Loan"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc=StandardScaler()
```

```python
x_train_new=sc.fit_transform(x_train)
```

```python
x_test_new=sc.transform(x_test)
```

```python
from sklearn.naive_bayes import GaussianNB
```

```python
gn=GaussianNB()
```

```python
gn.fit(x_train_new,y_train)
```

```python
y_pred=gn.predict(x_test_new)
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score , confusion_matrix 
```

```python
accuracy_score(y_test,y_pred)
```

```python
confusion_matrix(y_test,y_pred)
```
