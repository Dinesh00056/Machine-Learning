# Decision Tree

```python
# A decision tree is a supervised machine learning model that is used for classification and regression task that's why it is considered as (CART)
# It work like a flow chart or a tree structure where each internal node represents a feature and each branch represent a decison or condition and
#each leaf node represents an output  or class label in the classification or in a continuous value in the regression 
```

```python
# A decision tree split the data into subsets based on the value of the input features and this process is repeated recursively forming a tree like structure
```

# How It Works

```python
# 1). It starts with the entire data than it choose the best feature to split the data based on the metrices like ---> gini-index,entrophy,Mse(mean squad erroe)
# than it will split the dataset into the smaller subset.than it will repeat the process for each child node until a stoping condition met
```

```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv("covid_toy.csv")
```

```python
df.head(2)
```

```python
df=df.dropna()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
lb=LabelEncoder()
```

```python
df["gender"]=lb.fit_transform(df["gender"])
df["cough"]=lb.fit_transform(df["cough"])
df["city"]=lb.fit_transform(df["city"])
df["has_covid"]=lb.fit_transform(df["has_covid"])
```

```python
df.head(2)
```

```python
x=df.drop(columns=["has_covid"])
y=df["has_covid"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dt= DecisionTreeClassifier()
```

```python
dt.fit(x_train,y_train)
```

```python
y_pred=dt.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test,y_pred)
```

```python
df=pd.read_csv(" insurance.csv")
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
lb = LabelEncoder()
df['smoker' ] = lb.fit_transform(df['smoker'])
df['region' ] = lb.fit_transform(df['region'])
df['sex' ] = lb.fit_transform(df['sex'])
```

```python
dt=DecisionTreeRegressor()
```

```python

dt.fit(x_train,y_train)
```

```python
y_pred = dt.predict(x_test)
```

```python
from sklearn.metrics import r2_score
```
