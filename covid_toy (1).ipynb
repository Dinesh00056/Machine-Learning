```python
import pandas as pd
import numpy as np 

df=pd.read_csv("C:\\Users\\hp\\Regex\\data files\\covid_toy.csv")

df.head() 
```

```python
df.isnull().sum() 
```

```python
from sklearn.impute import SimpleImputer

si=SimpleImputer()

df['fever']=si.fit_transform(df[['fever']]) 
```

```python
df.isnull().sum() 
```

```python
df.head() 
```

```python
from sklearn.preprocessing import LabelEncoder

le=LabelEncoder() 

df['gender']=le.fit_transform(df['gender']) 
df['cough']=le.fit_transform(df['cough']) 
df['city']=le.fit_transform(df['city']) 
df['has_covid']=le.fit_transform(df['has_covid'])  
```

```python
df.head() 
```

```python
import tensorflow as tf
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder , StandardScaler
from sklearn.metrics import accuracy_score 
```

```python
x = df.drop(columns = ['has_covid'] , axis=1)
y = df['has_covid'] 
```

```python
x_train , x_test , y_train , y_test = train_test_split(x ,y ,test_size  = 0.2 ,random_state = 42) 
```

```python
sc = StandardScaler()
```

```python
x_train_scaled = sc.fit_transform(x_train)
x_test_scaled = sc.transform(x_test) 
```

```python
model = tf.keras.models.Sequential([
    tf.keras.layers.Dense(units = 128 , activation = 'relu' ,
                          input_dim = x_train_scaled.shape[1]),  #input layer
    tf.keras.layers.Dropout(0.3),
    tf.keras.layers.Dense(units = 64 , activation = 'relu'),    #hidden layer
    tf.keras.layers.Dropout(0.3),
    tf.keras.layers.Dense(units = 1 , activation = 'sigmoid')   #output layer
])
```

```python
loss_fn = tf.keras.losses.BinaryCrossentropy()
metrics = ['accuracy'] 
```

```python
learning_rate = 0.001
momentum = 0.9
optimizer = tf.keras.optimizers.SGD(learning_rate = learning_rate ,
                                    momentum = momentum , nesterov = True)
```

```python
model.compile(optimizer = optimizer , loss = loss_fn , metrics = metrics)
```

```python
model.fit(x_train_scaled ,
         y_train ,
         epochs = 50 ,
          batch_size = 8 ,
          validation_split = 0.1

         )
```

```python
loss , accuracy = model.evaluate(x_test_scaled , y_test)
```

```python
print("Total Accuracy :" , accuracy) 
```
