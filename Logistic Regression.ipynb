```python
# Logistic Regression----> ML Model ----> when we have target data as categorical
# In backend ---> Activation function ----> Sigmoid --> probability
# we know that probability lies from 0 to 1
# if prob > 0.5 or 1/2 ---> Output ---> 1
# if prob < 0.5 or 1.2 ----> output ---> 0
```

```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("covid_toy.csv")
```

```python
df.head(2)
```

```python
df.isnull().sum()
```

```python
from sklearn.impute import SimpleImputer
```

```python
si=SimpleImputer()
```

```python
df["fever"]=si.fit_transform(df[["fever"]])
```

```python
df.isnull().sum()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
lb=LabelEncoder()
```

```python
df["gender"]=lb.fit_transform(df["gender"])
df["cough"]=lb.fit_transform(df["cough"])
df["city"]=lb.fit_transform(df["city"])
df["has_covid"]=lb.fit_transform(df["has_covid"])
```

```python
x=df.drop(columns=["has_covid"])
y=df["has_covid"]
```

```python

from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr=LogisticRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
y_pred=lr.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test,y_pred)
```

```python
# confusion_metrix --- >
# TP --- > HUMAN DETECT AND MACHINE DETECT
# TN --- > HUMAN NOT DETECT AND MACHINE NOT DETECT
# FP --- > HUMAN NOT DETECT BUT MACHINE DETECT
# FN --- > HUMAN DETECT BUT MACHINE NOT DETECT
```

```python
from sklearn.metrics import confusion_matrix
```

```python
confusion_matrix(y_test,y_pred)
```
