# How Decision Tree Regressor works
The algorithm starts at the root node and recursively splits the data into the smaller subsets based on the specific features and on there values
the goal of each split is to create a child nodes that are more homogeneous in terms of the target variables to determine the best split at each node 
the algorithm typically uses a matrices like mean squarred error and mean absolute error. it evaluates potential split points for each feature and chooses 
the split that minimise the error in the resulting child nodes this process aims to reduces the variant of the target variable with a each subset

```python
import pandas as pd
```

```python
df=pd.read_csv("insurance.csv")
```

```python
df.head(2)
```

```python
df.isnull().sum()
```

```python
# df = pd.get_dummies(df , columns = ['sex','smoker','region'])
df=pd.get_dummies(df,columns=["sex",'smoker','region'])
```

```python
df.head(1)
```

```python
df.astype(int)
```

```python
x=df.drop(columns=["charges"])
y=df["charges"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
dt=DecisionTreeRegressor()
```

```python
dt.fit(x_train,y_train)
```

```python
y_pred=dt.predict(x_test)
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y_test , y_pred)
```
