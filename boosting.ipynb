```python
# 1) Adaboost -> It assigns higher weight to misclassifies existences nad focus on those instance during subsequential iteration it sequential trains a series of weak learners
# and combines there predictions to form the final assemble adaboost is used for binary classification problem 
# 2) Gradient Boosting -> This boosting builds an ensemble of week learners in a stage wise manner each subsequent model is trained to correct the mistakes 
#     made by the previous models by fitting the negative gradient of a loss function gradient boosting can handle both classification and regression task
#     and is often used with decision tree as a based learner 
#     Parts of Gradient Boosting
#   i)  XG boost -> XG boost is an optimised implimentation of gradient boosting that offers several inhancement like including parallel orocessing regularization
#     techniques and handling missing values. It uses a combination of tree based models and linear model for boosting which allows it to capture both linear and
#    non lonear relationship in data efficinetly
#    ii) LightGBM (Light Gradient Boosting Machine)-> light gradinet boosting framework that focus on achieving faster trained speed and lower memory uses.it uses a novel tree algorithm
#    called goss(gradinet based one side sampling) to select the most informative instances for building decision tree 
# iii) Cataboost -> cataboost is a gradient based algorithm that is designed to handle categorical features directly without the need for extensive data 
#         preprocessing It incorporates an inovative methods to handle categorical variables which includes applying a combination of ordered boosting 
#         random permutation and symmetric tree 
# iv)Stochastic Gradient Boosting->Stochastic Gradient Boosting boosting introduces randomness into the boosting process by subsampling the training data or features 
# at each iteration. It helps to reduce overfitting and can improve the model generalisation ability specially when dealing with large dataset
```

```python
from sklearn.ensemble import AdaBoostClassifier
from xgboost import XGBClassifier
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
#Generating a synthetic classification dataset
x,y = make_classification(n_samples=1000,n_features=10,random_state=42)
```

```python
# Splitting the dataset into training and testing dataset
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
#Creating and training Adaboost classifier
adaboost=AdaBoostClassifier(n_estimators=100,random_state=42)
adaboost.fit(x_train,y_train)
```

```python
#creating and traing an XGboost classifier
xgboost=XGBClassifier(n_estimators=100,random_state=42)
xgboost.fit(x_train,y_train)
```

```python
#Making predictions on the test set for adaboost
y_pred_adaboost=adaboost.predict(x_test)
```

```python
#Making predictions on the test set for XGboost
y_pred_xgboost=xgboost.predict(x_test)
```

```python
#calculating the accuracy of adaboost
accuracy_adaboost= accuracy_score(y_test,y_pred_adaboost)
accuracy_xgboost= accuracy_score(y_test,y_pred_xgboost)
```

```python
print("accuracy score",accuracy_adaboost)
print("accuracy score",accuracy_xgboost)
```

```python
# Importing necessary librories
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.ensemble import GradientBoostingClassifier
from lightgbm import LGBMClassifier
from catboost import CatBoostClassifier
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.ensemble import HistGradientBoostingClassifier
from sklearn.metrics import accuracy_score
```

```python
#Gradient Boosting Classifier
gb_classifier=GradientBoostingClassifier(n_estimators=100,random_state=42)
gb_classifier.fit(x_train,y_train)
y_pred_gb=gb_classifier.predict(x_test)
accuracy_gb= accuracy_score(y_test,y_pred_gb)
print("Gradinet Boosting Classifier Accuracy:",accuracy_gb)
```

```python
#LIghtGBM Classifier
lgb_classifier=GradientBoostingClassifier(n_estimators=100,random_state=42)
lgb_classifier.fit(x_train,y_train)
y_pred_lgb=lgb_classifier.predict(x_test)
accuracy_lgb= accuracy_score(y_test,y_pred_lgb)
print("LightGBM Classifier Accuracy:",accuracy_gb)
```

```python
 # CatBoost Classifier
cat_classifier=CatBoostClassifier(n_estimators=100,random_state=42)
cat_classifier.fit(x_train,y_train)
y_pred_cat=cat_classifier.predict(x_test)
accuracy_cat= accuracy_score(y_test,y_pred_cat)
print(" CatBoost Classifier Accuracy:",accuracy_cat)
```

```python
#Stochastic Gradinet Boosting Classifier
stoch_gb_classifier=HistGradientBoostingClassifier(max_iter=100,random_state=42)
stoch_gb_classifier.fit(x_train,y_train)
y_pred_stoch_gb=stoch_gb_classifier.predict(x_test)
accuracy_stoch_gb=accuracy_score(y_test,y_pred_stoch_gb)
print("Stochastic Gradinet Boosting Classifier accuracy:",accuracy_stoch_gb)
```

```python
#Gradient Boosting Regressor(for demonstration purposes)
gb_regressor=GradientBoostingRegressor(n_estimators=100,random_state=42)
gb_regressor.fit(x_train,y_train)
y_pred_gb_regressor=gb_regressor.predict(x_test)
```
