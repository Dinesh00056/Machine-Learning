```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("Social_Network_Ads - Social_Network_Ads.csv")
```

```python
df.head(1)
```

```python
df.isnull().sum()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
lb=LabelEncoder()
```

```python
df["Gender"]=lb.fit_transform(df["Gender"])
```

```python
x=df.drop(columns=["Purchased"])
y=df["Purchased"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
y_train.value_counts()
```

```python
from sklearn.ensemble import AdaBoostClassifier
from xgboost import XGBClassifier
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
#Generating a synthetic classification dataset
x,y = make_classification(n_samples=1000,n_features=10,random_state=42)
```

```python
#Creating and training Adaboost classifier
adaboost=AdaBoostClassifier(n_estimators=100,random_state=42)
adaboost.fit(x_train,y_train)
```

```python
#creating and traing an XGboost classifier
xgboost=XGBClassifier(n_estimators=100,random_state=42)
xgboost.fit(x_train,y_train)
```

```python
# Making predictions on the test set for adaboost
y_pred_adaboost=adaboost.predict(x_test)
```

```python
#Making predictions on the test set for XGboost
y_pred_xgboost=xgboost.predict(x_test)
```

```python
#calculating the accuracy of adaboost
accuracy_adaboost= accuracy_score(y_test,y_pred_adaboost)
accuracy_xgboost= accuracy_score(y_test,y_pred_xgboost)
```

```python
print("accuracy score",accuracy_adaboost)
print("accuracy score",accuracy_xgboost)
```

```python
from sklearn.datasets import make_classification
from sklearn.ensemble import GradientBoostingClassifier
from lightgbm import LGBMClassifier
from catboost import CatBoostClassifier
from sklearn.ensemble import HistGradientBoostingClassifier
```

```python
#GradientBoostingClassifier
gb_classifier=GradientBoostingClassifier(n_estimators=100,random_state=42)
gb_classifier.fit(x_train,y_train)
y_pred=gb_classifier.predict(x_test)
accuracy_gb=accuracy_score(y_test,y_pred)
print("GradientBoostingClassifier",accuracy_gb)
```

```python
# LightGBM Classifier
lgb_classifier = GradientBoostingClassifier(n_estimators=100,random_state=42)
lgb_classifier.fit(x_train,y_train)
y_pred_lgb=lgb_classifier.predict(x_test)
accuracy_lgb=accuracy_score(y_test,y_pred_lgb)
print("LightGBM Classifier Accuracy:",accuracy_gb)
```

```python
# CatBoost Classifier
cat_classifier=CatBoostClassifier(n_estimators=100,random_state=42)
cat_classifier.fit(x_train,y_train)
y_pred_cat=cat_classifier.predict(x_test)
accuracy_cat=accuracy_score(y_test,y_pred_cat)
print("CatBoost Classifier Accuracy:",accuracy_cat)
```

```python
#Stochastic Gradinet Boosting Classifier
stoch_gb_classifier=HistGradientBoostingClassifier(max_iter=100,random_state=42)
stoch_gb_classifier.fit(x_train,y_train)
y_pred_stoch_gb=stoch_gb_classifier.predict(x_test)
accuracy_stoch_gb=accuracy_score(y_test,y_pred_stoch_gb)
print("Stochastic Gradinet Boosting Classifier accuracy:",accuracy_stoch_gb)
```
