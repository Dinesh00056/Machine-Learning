```python
# Gradient Descent Optimization Technique----------->
```

```python
# Loss function => Actual_value - Predicted_value
# Optimization => Reduce the loss function.
    
```

```python
# #Let suppose, if we have equation of simple line:
# y= mx+b .....(1) (x,y)---> coordinates , m= slope(weights),b=intercept
```

```python
# now let suppose,prediction line is :
# y_pred = mx+b ....(2)
```

```python
# Now, we will calculate the loss function with respect to mean square error
# Loss= 1/n * (y-y_pred)**2
# n= n(features)=1
# Loss= (y-y_pred)**2
# Loss= (y-(mx+b))**2 ....(3)
```

```python
#  Now we will calculate the differentiation with respect to intercept(b):
# dl/db =2*(y-mx-b)[dy/db-dmx/dm-db/dm]
# dl/dm = 2*(y-mx-b)[0-x-0]
# dl/dm = -2(y-mx-b)(x).....(5)
```

```python
#  In general lr = 0.001
# b_new=b_old - learning_rate*dl/db
# m_new=m_old - learning_rate *dl/dm
```

```python
# b_old=50 , dl/db=100 = b_new=50-0.0001*100 = 50-0.1 = 49.9
```

```python
#lr>>>>>0 ---> gradient exploading
#lr<<<<0 ---> vanishing gradient
```

```python
# # So , our final line of equation will be :
# y_new = m_new*X+b_new   ..........(6)
```

```python
# Equation(6 denotes as a final line of equation)
```
