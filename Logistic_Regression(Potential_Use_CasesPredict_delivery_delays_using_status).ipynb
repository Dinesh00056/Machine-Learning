```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("/content/logistics_shipments_dataset - logistics_shipments_dataset.csv")
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
from sklearn.impute import SimpleImputer
```

```python
si=SimpleImputer(strategy='most_frequent')
```

```python
df[["Delivery_Date","Cost"]]=si.fit_transform(df[["Delivery_Date","Cost"]])
```

```python
df.isnull().sum()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
lb=LabelEncoder()
```

```python
df.head(1)
```

```python
df["Shipment_ID"]=lb.fit_transform(df["Shipment_ID"])
df["Origin_Warehouse"]=lb.fit_transform(df["Origin_Warehouse"])
df["Destination"]=lb.fit_transform(df["Destination"])
df["Carrier"]=lb.fit_transform(df["Carrier"])
df["Shipment_Date"]=lb.fit_transform(df["Shipment_Date"])
df["Status"]=lb.fit_transform(df["Status"])
df["Delivery_Date"]=lb.fit_transform(df["Delivery_Date"])
```

```python
x=df.drop(columns=["Status"])
y=df["Status"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr=LogisticRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
y_pred=lr.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test,y_pred)
```

```python
param_grid = [
    {'penalty':['l1','l2','elasticnet','none'],
    'C' : np.logspace(-4,4,20),
    'solver' : ['lbfgs', 'newton-cg','liblinear','sag','saga'],
    'max_iter' : [10,50,100,1000]

}
]
```

```python
from sklearn.model_selection import GridSearchCV
clf = GridSearchCV(lr,param_grid= param_grid,cv=4,verbose=True,n_jobs=-1)
```

```python
best_clf=clf.fit(x,y)
best_clf.best_estimator_
```

```python
print(f'accuracy-:{best_clf.score(x,y):.3f}')
```
