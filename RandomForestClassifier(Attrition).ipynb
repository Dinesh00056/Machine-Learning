```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("Attrition.csv")
```

```python
df.head(1)
```

```python
df.dtypes
```

```python
df.isnull().sum()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
lb=LabelEncoder()
```

```python
df.head(1)
```

```python
df["Attrition"]=lb.fit_transform(df["Attrition"])
df["BusinessTravel"]=lb.fit_transform(df["BusinessTravel"])
df["Department"]=lb.fit_transform(df["Department"])
df["EducationField"]=lb.fit_transform(df["EducationField"])
df["Gender"]=lb.fit_transform(df["Gender"])
df["JobRole"]=lb.fit_transform(df["JobRole"])
df["MaritalStatus"]=lb.fit_transform(df["MaritalStatus"])
df["Over18"]=lb.fit_transform(df["Over18"])
df["OverTime"]=lb.fit_transform(df["OverTime"])
```

```python
df.dtypes
```

```python
x=df.drop(columns=["Attrition"])
y=df["Attrition"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc=StandardScaler()
```

```python
x_train_new=sc.fit_transform(x_train)
```

```python
x_test_new=sc.transform(x_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf=RandomForestClassifier()
```

```python
rf.fit(x_train_new,y_train)
```

```python
y_pred=rf.predict(x_test_new)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test,y_pred)
```

```python
from imblearn.over_sampling import SMOTE
```

```python
smote = SMOTE(random_state=42)
x_train_balanced, y_train_balanced = smote.fit_resample(x_train, y_train)
```

```python
rf = RandomForestClassifier(n_estimators=100, random_state=42)
rf.fit(x_train_balanced, y_train_balanced)
```

```python
y_pred = rf.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test,y_pred)
```

```python
from sklearn.model_selection import GridSearchCV, RandomizedSearchCV
```

```python
param_grid={
    "n_estimators":[10,20,30],
    "max_depth":[None,10,20,30],
    "min_samples_split":[2,5,8],
    "min_samples_leaf":[1,2,4],
    "bootstrap":[True,False]
}
```

```python
grid_search=GridSearchCV(RandomForestClassifier(),param_grid=param_grid,cv=5)
grid_search.fit(x_train,y_train)
```

```python
print("Best Paramater:",grid_search.best_params_)
print("Best Estimator:",grid_search.best_estimator_)
```

```python
model=RandomForestClassifier(max_depth=3,
                            max_features="log2",
                            max_leaf_nodes=3,
                            n_estimators=50)
model.fit(x_train,y_train)
y_pred= model.predict(x_test)
print(accuracy_score(y_pred,y_test))
```
